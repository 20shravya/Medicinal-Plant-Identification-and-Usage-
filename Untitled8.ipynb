```python
# Step 3 вЂ” list misclassified Sandalwood val images (and predicted class)
import json, numpy as np, tensorflow as tf
from pathlib import Path
from PIL import Image

BASE = Path(r"C:\MedicinalPlantApp")
MODEL_PATH = BASE/"models"/"plant_model_final.h5"   # or plant_model_after_sandal_fix.h5 if you saved a new one
model = tf.keras.models.load_model(str(MODEL_PATH))

with open(BASE/"models"/"class_indices.json") as f:
    class_indices = json.load(f)
idx_to_class = {v:k for k,v in class_indices.items()}

val_dir = BASE/"data"/"val"/"Sandalwood"
mis = []
for p in sorted(val_dir.glob("*.*")):
    try:
        img = Image.open(p).convert("RGB").resize((160,160))
        x = np.expand_dims(np.asarray(img)/255.0,0)
        preds = model.predict(x)[0]
        top = int(preds.argmax())
        top_class = idx_to_class[top]
        if top_class != "Sandalwood":
            mis.append((p.name, top_class, float(preds[top])))
    except Exception as e:
        print("ERROR reading:", p.name, e)

print("Found", len(mis), "misclassified Sandalwood images (showing up to 100):")
for it in mis[:100]:
    print(it)
```

```python
# STEP 2 вЂ” Augment extra Sandalwood images into train/Sandalwood
from pathlib import Path
import shutil, os
from tensorflow.keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
import random

BASE = Path(r"C:\MedicinalPlantApp")
EXTRA = BASE/"data"/"extra_sandalwood"
TRAIN_SAND = BASE/"data"/"train"/"Sandalwood"

# ensure directory layout
if not EXTRA.exists():
    raise SystemExit(f"Extra folder missing: {EXTRA} вЂ” please create and copy 50-100 example images there and re-run.")
# if images are in EXTRA root, move them into EXTRA/Sandalwood
if not (EXTRA/"Sandalwood").exists():
    (EXTRA/"Sandalwood").mkdir(parents=True, exist_ok=True)
    moved = 0
    for f in sorted(EXTRA.glob("*.*")):
        if f.is_file() and f.suffix.lower() in [".png",".jpg",".jpeg",".bmp",".tiff"]:
            # skip files already inside the subfolder
            if f.parent != (EXTRA/"Sandalwood"):
                shutil.move(str(f), str(EXTRA/"Sandalwood"/f.name))
                moved += 1
    if moved:
        print(f"Moved {moved} images into {EXTRA/'Sandalwood'}")

# confirm sources
src_files = list((EXTRA/"Sandalwood").glob("*.*"))
if len(src_files) == 0:
    raise SystemExit("No images found inside extra_sandalwood/Sandalwood вЂ” copy 50-100 images there and re-run.")

print("Found", len(src_files), "source images for augmentation.")

# augmentation settings вЂ” strong but reasonable
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.20,
    height_shift_range=0.20,
    shear_range=0.15,
    zoom_range=0.20,
    brightness_range=(0.6,1.4),
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest'
)

TRAIN_SAND.mkdir(parents=True, exist_ok=True)

# count how many already in train
existing = list(TRAIN_SAND.glob("*.*"))
existing_count = len(existing)
print("Existing in train/Sandalwood:", existing_count)

# how many augmented images to create
n_to_generate = 400   # <--- change if you want more/less
print("Will generate", n_to_generate, "augmented images into", TRAIN_SAND)

gen_count = 0
src_list = src_files.copy()
i = 0
# create augmented images one-by-one (keeps filenames unique)
while gen_count < n_to_generate:
    src = random.choice(src_list)
    img = load_img(src, target_size=(160,160))  # resize for augmentation
    x = img_to_array(img)
    x = x.reshape((1,) + x.shape)  # (1, h, w, c)
    prefix = f"aug_{i}_"
    for batch in datagen.flow(x, batch_size=1, save_to_dir=str(TRAIN_SAND),
                              save_prefix=prefix, save_format='png'):
        gen_count += 1
        i += 1
        if gen_count % 50 == 0:
            print(f"Generated {gen_count}/{n_to_generate}")
        break  # only one generated per iteration so we can control count

print("Augmentation complete. New images generated:", gen_count)
print("Final count in train/Sandalwood:", len(list(TRAIN_SAND.glob("*.*"))))
```

```python
# STEP 3 вЂ” Short fine-tune using augmented Sandalwood
import json, numpy as np, tensorflow as tf
from pathlib import Path
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models, optimizers
from sklearn.utils.class_weight import compute_class_weight

BASE = Path(r"C:\MedicinalPlantApp")
DATA_DIR = BASE/"data"
MODELS_DIR = BASE/"models"
MODELS_DIR.mkdir(exist_ok=True)

IMG_SIZE = 160
BATCH = 16
FT_EPOCHS = 6

# prepare generators (light augmentation for training)
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=15,
                                   width_shift_range=0.08,
                                   height_shift_range=0.08,
                                   zoom_range=0.08,
                                   horizontal_flip=True)

val_datagen = ImageDataGenerator(rescale=1./255)

train_flow = train_datagen.flow_from_directory(
    str(DATA_DIR/"train"),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH,
    class_mode='categorical'
)
val_flow = val_datagen.flow_from_directory(
    str(DATA_DIR/"val"),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH,
    class_mode='categorical',
    shuffle=False
)

# compute class weights
with open(MODELS_DIR/"class_indices.json") as f:
    class_indices = json.load(f)   # name -> idx
# build y array from filenames in train directory
y = []
idx_to_name = {v:k for k,v in class_indices.items()}
for cname, idx in class_indices.items():
    p = DATA_DIR/"train"/cname
    if p.exists():
        # count files
        cnt = len([x for x in p.glob("*.*") if x.is_file()])
        y += [idx]*cnt

y = np.array(y, dtype=np.int32)
classes = np.unique(y)
cw = compute_class_weight("balanced", classes=classes, y=y)
class_weight = {int(classes[i]): float(cw[i]) for i in range(len(classes))}
print("Sample class_weight items:", list(class_weight.items())[:6])

# load model (best final) вЂ” change filename if you used a different one
model_path = MODELS_DIR/"plant_model_final.h5"
model = tf.keras.models.load_model(str(model_path))
print("Loaded model:", model_path)

# unfreeze last layers of base model if using MobileNetV2
# find backbone by name or layer count вЂ” we'll unfreeze last 30 layers if available
base = None
for layer in model.layers:
    # find first Conv2D after input as heuristic for base
    if hasattr(layer, "name") and layer.name.startswith("Conv1"):
        # not used; instead rely on layer list
        pass

# Make last ~30 layers trainable
for layer in model.layers:
    layer.trainable = False
# try to unfreeze last 30 layers safely
unfreeze_n = 30
if len(model.layers) > unfreeze_n:
    for layer in model.layers[-unfreeze_n:]:
        layer.trainable = True
else:
    for layer in model.layers:
        layer.trainable = True

# compile with low lr
model.compile(optimizer=optimizers.Adam(1e-5),
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# callbacks
cb = [
    tf.keras.callbacks.ModelCheckpoint(str(MODELS_DIR/"plant_model_sandal_fix.h5"), save_best_only=True),
    tf.keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=2, min_lr=1e-7)
]

# run fine-tune
history = model.fit(
    train_flow,
    validation_data=val_flow,
    epochs=FT_EPOCHS,
    class_weight=class_weight,
    callbacks=cb
)

# save final tuned model
model.save(MODELS_DIR/"plant_model_after_sandal_fix.h5")
print("Saved tuned model exists:", (MODELS_DIR/"plant_model_after_sandal_fix.h5").exists())

# print last epoch results
if "accuracy" in history.history:
    print("Last train acc:", history.history["accuracy"][-1])
if "val_accuracy" in history.history:
    print("Last val acc:", history.history["val_accuracy"][-1])
```
