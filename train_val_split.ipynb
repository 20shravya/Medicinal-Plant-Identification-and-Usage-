```python
import os, shutil, random
from pathlib import Path

data_path = Path(r"C:\MedicinalPlantApp\data")
train_dir = data_path / "train"
val_dir = data_path / "val"

val_dir.mkdir(exist_ok=True)

for cls in train_dir.iterdir():
    if cls.is_dir():
        images = list(cls.glob("*"))
        random.shuffle(images)
        val_count = max(1, int(len(images) * 0.20))  # 20%

        (val_dir / cls.name).mkdir(parents=True, exist_ok=True)

        for img in images[:val_count]:
            shutil.move(str(img), str(val_dir / cls.name / img.name))

print("Validation split created successfully!")
```
