```python
# Single-image prediction — replace test_path with one path you copied and run
from tensorflow.keras.preprocessing import image
import numpy as np
import json
from pathlib import Path
import tensorflow as tf
from IPython.display import display
from PIL import Image

BASE = Path(r"C:\MedicinalPlantApp")
MODEL_PATH = BASE/"models"/"plant_model.h5"

# load model + mapping
model = tf.keras.models.load_model(str(MODEL_PATH))
with open(BASE/"models"/"class_indices.json") as f:
    class_indices = json.load(f)
idx_to_class = {v:k for k,v in class_indices.items()}

# <-- REPLACE this with one full path you copied above
test_path = r"C:\MedicinalPlantApp\data\val\Arive-Dantu\__100_1138909.png"

# load + preprocess
img = Image.open(test_path).convert("RGB")
img_resized = img.resize((160,160))
display(img)  # shows the image in notebook

x = image.img_to_array(img_resized) / 255.0
x = np.expand_dims(x, 0)

pred = model.predict(x)[0]
top3 = np.argsort(pred)[-3:][::-1]

print("\nTop predictions:")
for i in top3:
    print(f"{idx_to_class[i]} — {pred[i]:.4f}")
print(f"\nPredicted class (top-1): {idx_to_class[top3[0]]} (confidence {pred[top3[0]]:.4f})")
```
