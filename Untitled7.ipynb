```python
# Step 1 — counts
from pathlib import Path
BASE = Path(r"C:\MedicinalPlantApp\data")
for split in ["train","val"]:
    p = BASE / split
    counts = {d.name: sum(1 for _ in d.rglob("*") if _.is_file()) for d in p.iterdir() if d.is_dir()}
    print(f"\n--- {split} counts (top 50) ---")
    for k,v in sorted(counts.items(), key=lambda x:-x[1])[:50]:
        print(f"{k:30s} {v:4d}")
    print("Sandalwood:", counts.get("Sandalwood", 0))
```

```python
# Step 2 — display Sandalwood examples
from pathlib import Path
from PIL import Image
from IPython.display import display

BASE = Path(r"C:\MedicinalPlantApp\data")
for split in ["train","val"]:
    sdir = BASE / split / "Sandalwood"
    print(f"\n{split} samples from {sdir}:")
    if not sdir.exists():
        print("  MISSING")
        continue
    imgs = sorted(sdir.glob("*.*"))[:12]
    print("  showing", len(imgs), "images")
    for imgp in imgs:
        print(" -", imgp.name)
        try:
            display(Image.open(imgp).resize((240,240)))
        except Exception as e:
            print("   ERROR opening:", imgp.name, e)
```
