```python
# STEP 3 — ROBUST: auto-detect model input size, ignore dot-folders, generate misclassified.csv
import os, shutil, csv, fnmatch
import numpy as np
import tensorflow as tf
from collections import Counter

# ========== CONFIG ==========
PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")
OUTPUT_CSV = os.path.join(PROJECT_ROOT, "misclassified.csv")
OUT_SANDAL_DIR = os.path.join(PROJECT_ROOT, "misclassified_sandalwood")
MAX_COPY = 200
# ==========================

def find_models(root):
    found = []
    for current, dirs, files in os.walk(root):
        for pattern in ("*.h5", "*.hdf5", "*.keras"):
            for fname in fnmatch.filter(files, pattern):
                found.append(os.path.join(current, fname))
        if "saved_model.pb" in files:
            found.append(current)
    return found

print("Searching for model files under:", PROJECT_ROOT)
models = find_models(PROJECT_ROOT)
if not models:
    raise SystemExit("No model file found under the project root. Place your .h5/.keras or SavedModel there.")

MODEL_PATH = models[0]
print("Using model:", MODEL_PATH)

# load model
model = tf.keras.models.load_model(MODEL_PATH)

# determine input size from model
try:
    spec = model.input_shape  # e.g. (None, 160, 160, 3) or (None, 224, 224, 3)
    if isinstance(spec, list):
        spec = spec[0]
    _, h, w, c = spec
    if h is None or w is None:
        raise ValueError
    IMG_SIZE = (int(h), int(w))
except Exception:
    # fallback
    IMG_SIZE = (224, 224)

print("Model input size detected:", IMG_SIZE)

# get validation classes, ignore hidden dot-folders
if not os.path.isdir(VAL_DIR):
    raise SystemExit(f"Validation folder not found: {VAL_DIR}")

classes = sorted([d for d in os.listdir(VAL_DIR)
                  if os.path.isdir(os.path.join(VAL_DIR, d)) and not d.startswith('.')])
print("Validation classes:", classes)

os.makedirs(OUT_SANDAL_DIR, exist_ok=True)

rows = []
sandal_confusions = []
total = 0
sandal_count = 0
copied = 0

for i, cls in enumerate(classes):
    cls_folder = os.path.join(VAL_DIR, cls)
    for fname in os.listdir(cls_folder):
        if not fname.lower().endswith((".jpg", ".jpeg", ".png")):
            continue
        total += 1
        path = os.path.join(cls_folder, fname)

        # load image with target size = model input
        img = tf.keras.preprocessing.image.load_img(path, target_size=IMG_SIZE)
        x = tf.keras.preprocessing.image.img_to_array(img) / 255.0
        x = np.expand_dims(x, axis=0)

        preds = model.predict(x, verbose=0)
        # if model outputs logits or probs, take argmax on last axis; handle unexpected shapes
        pred_idx = int(np.argmax(preds, axis=-1)[0])
        pred_cls = classes[pred_idx] if pred_idx < len(classes) else "UNKNOWN"

        if pred_idx != i:
            rows.append([path, cls, pred_cls])
            if cls.lower() == "sandalwood":
                sandal_confusions.append(pred_cls)
                sandal_count += 1
                if copied < MAX_COPY:
                    newname = f"true_{cls}__pred_{pred_cls}__{fname}"
                    try:
                        shutil.copy(path, os.path.join(OUT_SANDAL_DIR, newname))
                        copied += 1
                    except Exception:
                        pass

# save CSV
with open(OUTPUT_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["path", "true", "pred"])
    writer.writerows(rows)

print("\n=== SUMMARY ===")
print("Model used:", MODEL_PATH)
print("Model input size used:", IMG_SIZE)
print("Total validation images:", total)
print("Total misclassified:", len(rows))
print("Total sandalwood misclassified:", sandal_count)

if sandal_count > 0:
    print("\nTop sandalwood confusions:")
    for k, v in Counter(sandal_confusions).most_common(10):
        print(f"  {k}: {v}")

print("\nCSV saved at:", OUTPUT_CSV)
print("Misclassified sandalwood images (copied) at:", OUT_SANDAL_DIR)
```

```python
# STEP 4 — Confusion matrix + classification report (robust)
import os, csv, itertools
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
from collections import Counter

PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")
MODEL_PATH = None  # will auto-find first .h5/.keras/SavedModel like before
OUT_CM_CSV = os.path.join(PROJECT_ROOT, "confusion_matrix.csv")
OUT_REPORT_CSV = os.path.join(PROJECT_ROOT, "classification_report.csv")
OUT_CM_PNG = os.path.join(PROJECT_ROOT, "confusion_matrix.png")

# auto-find model (first found)
import fnmatch
def find_models(root):
    found = []
    for current, dirs, files in os.walk(root):
        for pattern in ("*.h5", "*.hdf5", "*.keras"):
            for fname in fnmatch.filter(files, pattern):
                found.append(os.path.join(current, fname))
        if "saved_model.pb" in files:
            found.append(current)
    return found

models = find_models(PROJECT_ROOT)
if not models:
    raise SystemExit("No model found in project root.")
MODEL_PATH = models[0]
print("Using model:", MODEL_PATH)

model = tf.keras.models.load_model(MODEL_PATH)

# infer input size
spec = model.input_shape
if isinstance(spec, list):
    spec = spec[0]
_, H, W, C = spec
if H is None or W is None:
    IMG_SIZE = (160, 160)
else:
    IMG_SIZE = (int(H), int(W))
print("Using image size:", IMG_SIZE)

# get classes (ignore dot-folders)
classes = sorted([d for d in os.listdir(VAL_DIR) if os.path.isdir(os.path.join(VAL_DIR, d)) and not d.startswith('.')])
print("Classes:", classes)

# collect predictions
y_true = []
y_pred = []
paths = []

for i, cls in enumerate(classes):
    cls_folder = os.path.join(VAL_DIR, cls)
    for fname in os.listdir(cls_folder):
        if not fname.lower().endswith((".jpg", ".jpeg", ".png")):
            continue
        p = os.path.join(cls_folder, fname)
        img = tf.keras.preprocessing.image.load_img(p, target_size=IMG_SIZE)
        x = tf.keras.preprocessing.image.img_to_array(img) / 255.0
        x = np.expand_dims(x, axis=0)
        preds = model.predict(x, verbose=0)
        pred_idx = int(np.argmax(preds, axis=-1)[0])
        y_true.append(i)
        y_pred.append(pred_idx)
        paths.append(p)

y_true = np.array(y_true)
y_pred = np.array(y_pred)

# classification report
report = classification_report(y_true, y_pred, target_names=classes, output_dict=True, zero_division=0)
# print nicely
print("\n=== Classification report (precision / recall / f1 / support) ===\n")
for cls in classes:
    pr = report[cls]
    print(f"{cls:20s}  precision: {pr['precision']:.4f}  recall: {pr['recall']:.4f}  f1: {pr['f1-score']:.4f}  support: {int(pr['support'])}")

# save report CSV
with open(OUT_REPORT_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["class","precision","recall","f1-score","support"])
    for cls in classes:
        pr = report[cls]
        writer.writerow([cls, pr['precision'], pr['recall'], pr['f1-score'], int(pr['support'])])
print("\nSaved classification report to:", OUT_REPORT_CSV)

# confusion matrix
cm = confusion_matrix(y_true, y_pred)
# save confusion matrix as CSV (rows = true classes)
with open(OUT_CM_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["true\\pred"] + classes)
    for i, row in enumerate(cm):
        writer.writerow([classes[i]] + row.tolist())
print("Saved confusion matrix to:", OUT_CM_CSV)

# show sandalwood confusions
if "Sandalwood" in classes:
    si = classes.index("Sandalwood")
    row = cm[si]
    conf_counts = {classes[j]: int(row[j]) for j in range(len(classes)) if row[j] > 0}
    sorted_conf = sorted(conf_counts.items(), key=lambda x: x[1], reverse=True)
    print("\nTop classes Sandalwood is confused with (true=Sandalwood):")
    for cls_name, cnt in sorted_conf[:10]:
        print(f"  {cls_name}: {cnt}")
else:
    print("\nWarning: 'Sandalwood' not found among classes (check spelling/case).")

# plot confusion matrix heatmap (normalized by true-row)
import matplotlib.pyplot as plt
import numpy as np
cm_norm = cm.astype('float') / (cm.sum(axis=1, keepdims=True) + 1e-8)

fig, ax = plt.subplots(figsize=(14, 12))
im = ax.imshow(cm_norm, interpolation='nearest', aspect='auto')
ax.set_title("Confusion matrix (normalized by true class)")
ax.set_xticks(np.arange(len(classes)))
ax.set_yticks(np.arange(len(classes)))
ax.set_xticklabels(classes, rotation=90, fontsize=8)
ax.set_yticklabels(classes, fontsize=8)
plt.colorbar(im, ax=ax, fraction=0.03)
# annotate a few cells with absolute counts for sandalwood row/top confusions
si = classes.index("Sandalwood") if "Sandalwood" in classes else None
if si is not None:
    for j in range(len(classes)):
        val = cm[si, j]
        if val>0:
            ax.text(j, si, str(val), ha="center", va="center", color="white", fontsize=7)

plt.tight_layout()
plt.savefig(OUT_CM_PNG, dpi=150)
print("Saved confusion matrix image to:", OUT_CM_PNG)
plt.show()
```
