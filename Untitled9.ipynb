```python
# FULL RETRAIN (all classes) — run from C:\MedicinalPlantApp in a NEW notebook cell
import json, numpy as np, tensorflow as tf
from pathlib import Path
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models, optimizers
from sklearn.utils.class_weight import compute_class_weight

BASE = Path(r"C:\MedicinalPlantApp")
DATA_DIR = BASE/"data"
MODELS_DIR = BASE/"models"
LOGS_DIR = BASE/"logs"
MODELS_DIR.mkdir(exist_ok=True)
LOGS_DIR.mkdir(exist_ok=True)

IMG_SIZE = 160
BATCH = 16          # lower to 8 if you run out of memory
EPOCHS_MAIN = 20    # main training
EPOCHS_FT = 5       # fine-tuning

print("DATA_DIR:", DATA_DIR)
print("Using IMG_SIZE, BATCH:", IMG_SIZE, BATCH)

# Data generators (stronger augmentation for training)
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=25,
                                   width_shift_range=0.12,
                                   height_shift_range=0.12,
                                   zoom_range=0.12,
                                   horizontal_flip=True,
                                   brightness_range=(0.8,1.2))
val_datagen = ImageDataGenerator(rescale=1./255)

train_flow = train_datagen.flow_from_directory(
    str(DATA_DIR/"train"),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH,
    class_mode='categorical'
)
val_flow = val_datagen.flow_from_directory(
    str(DATA_DIR/"val"),
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH,
    class_mode='categorical',
    shuffle=False
)

# Save class indices (overwrite if needed)
with open(MODELS_DIR/"class_indices.json", "w") as f:
    json.dump(train_flow.class_indices, f)
print("Saved class indices. Num classes:", len(train_flow.class_indices))

# compute class weights from train folder counts
class_indices = train_flow.class_indices  # name -> idx
# count per class
counts = {}
for cname, idx in class_indices.items():
    p = DATA_DIR/"train"/cname
    counts[cname] = len([x for x in p.glob("*.*") if x.is_file()]) if p.exists() else 0
# build y array
y = []
for cname, idx in class_indices.items():
    y += [idx] * counts.get(cname, 0)
y = np.array(y, dtype=np.int32)
classes = np.unique(y)
if len(classes) == 0:
    raise SystemExit("No training files found! Check DATA_DIR/train.")
cw = compute_class_weight("balanced", classes=classes, y=y)
class_weight = {int(classes[i]): float(cw[i]) for i in range(len(classes))}
print("Sample class_weight:", list(class_weight.items())[:6])

# Build model (transfer learning)
base = tf.keras.applications.MobileNetV2(weights="imagenet", include_top=False, input_shape=(IMG_SIZE,IMG_SIZE,3))
base.trainable = False
x = base.output
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(256, activation="relu")(x)
outputs = layers.Dense(len(train_flow.class_indices), activation="softmax")(x)
model = models.Model(inputs=base.input, outputs=outputs)

model.compile(optimizer=optimizers.Adam(1e-3),
              loss="categorical_crossentropy",
              metrics=["accuracy"])
model.summary()

# Callbacks for main training
callbacks_main = [
    tf.keras.callbacks.ModelCheckpoint(str(MODELS_DIR/"plant_model_best_main.h5"), save_best_only=True),
    tf.keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=3, min_lr=1e-6),
    tf.keras.callbacks.TensorBoard(log_dir=str(LOGS_DIR), histogram_freq=0)
]

# Main training
print("\n=== START main training (epochs =", EPOCHS_MAIN, ") ===")
history_main = model.fit(
    train_flow,
    validation_data=val_flow,
    epochs=EPOCHS_MAIN,
    callbacks=callbacks_main,
    class_weight=class_weight
)
print("=== MAIN training finished ===")

# Fine-tuning: unfreeze last layers of the base
print("\n=== START fine-tuning ===")
base.trainable = True
# freeze all but last N layers
N_unfreeze = 30
for layer in base.layers[:-N_unfreeze]:
    layer.trainable = False
for layer in base.layers[-N_unfreeze:]:
    layer.trainable = True

model.compile(optimizer=optimizers.Adam(1e-5),
              loss="categorical_crossentropy",
              metrics=["accuracy"])

callbacks_ft = [
    tf.keras.callbacks.ModelCheckpoint(str(MODELS_DIR/"plant_model_best_ft.h5"), save_best_only=True),
    tf.keras.callbacks.EarlyStopping(patience=4, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(factor=0.5, patience=2, min_lr=1e-7)
]

history_ft = model.fit(
    train_flow,
    validation_data=val_flow,
    epochs=EPOCHS_FT,
    callbacks=callbacks_ft,
    class_weight=class_weight
)
print("=== FINE-TUNING finished ===")

# Save final model
final_path = MODELS_DIR/"plant_model_full.h5"
model.save(final_path)
print("Saved final model:", final_path.exists())

# Print summary results
if "accuracy" in history_main.history:
    print("Main training last train acc:", history_main.history["accuracy"][-1])
if "val_accuracy" in history_main.history:
    print("Main training last val acc:", history_main.history["val_accuracy"][-1])
if "accuracy" in history_ft.history:
    print("Fine-tune last train acc:", history_ft.history["accuracy"][-1])
if "val_accuracy" in history_ft.history:
    print("Fine-tune last val acc:", history_ft.history["val_accuracy"][-1])
```
