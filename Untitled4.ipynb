```python
import os, fnmatch, numpy as np, tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array

PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")

# find first model file
models = []
for root, dirs, files in os.walk(PROJECT_ROOT):
    for pattern in ("*.h5", "*.hdf5", "*.keras"):
        for fname in fnmatch.filter(files, pattern):
            models.append(os.path.join(root, fname))
models = sorted(models)
print("Models found (top 5):", models[:5])
if not models:
    raise SystemExit("No model files found under project root.")

MODEL_PATH = models[0]
print("Loading model:", MODEL_PATH)
model = tf.keras.models.load_model(MODEL_PATH)
print("Model loaded. model.input_shape =", model.input_shape)

# choose a test image (prefer Sandalwood)
classes = sorted([d for d in os.listdir(VAL_DIR) if os.path.isdir(os.path.join(VAL_DIR, d)) and not d.startswith('.')])
test_img = None
for cls in classes:
    imgs = [f for f in os.listdir(os.path.join(VAL_DIR, cls)) if f.lower().endswith((".jpg",".jpeg",".png"))]
    if imgs:
        test_img = os.path.join(VAL_DIR, cls, imgs[0])
        break

print("Test image selected:", test_img)
H = model.input_shape[1] or 160
W = model.input_shape[2] or 160
img = load_img(test_img, target_size=(H,W))
x = img_to_array(img)/255.0
x = np.expand_dims(x, axis=0)
preds = model.predict(x, verbose=0)
print("Prediction shape:", np.shape(preds))
print("Top 6 prediction values (first 6):", preds.ravel()[:6])
print("Argmax index:", int(preds.argmax()))

```

```python
# Diagnostic 3 — quick misclassification summary (prints progress)
import os, fnmatch, numpy as np
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from collections import Counter
import tensorflow as tf

PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")

# load first model found (reuse if already in memory)
models = []
for root, dirs, files in os.walk(PROJECT_ROOT):
    for pattern in ("*.h5", "*.hdf5", "*.keras"):
        for fname in fnmatch.filter(files, pattern):
            models.append(os.path.join(root, fname))
models = sorted(models)
MODEL_PATH = models[0]
print("Using model:", MODEL_PATH)
model = tf.keras.models.load_model(MODEL_PATH)
print("Model input shape:", model.input_shape)

classes = sorted([d for d in os.listdir(VAL_DIR) if os.path.isdir(os.path.join(VAL_DIR, d)) and not d.startswith('.')])
IMG_H = model.input_shape[1] or 160
IMG_W = model.input_shape[2] or 160

y_true = []
y_pred = []
total = 0
for i, cls in enumerate(classes):
    cls_folder = os.path.join(VAL_DIR, cls)
    for fname in os.listdir(cls_folder):
        if not fname.lower().endswith((".jpg", ".jpeg", ".png")): continue
        p = os.path.join(cls_folder, fname)
        img = load_img(p, target_size=(IMG_H, IMG_W))
        x = img_to_array(img)/255.0
        x = np.expand_dims(x, axis=0)
        pred_idx = int(model.predict(x, verbose=0).argmax())
        y_true.append(i); y_pred.append(pred_idx)
        total += 1
        if total % 200 == 0:
            print("Processed", total, "images...")
y_true = np.array(y_true); y_pred = np.array(y_pred)
mis = int((y_true != y_pred).sum())
sandal_idx = classes.index("Sandalwood") if "Sandalwood" in classes else None
sandal_mis = int(((y_true==sandal_idx) & (y_true!=y_pred)).sum()) if sandal_idx is not None else "N/A"
print("\nDone. Total images processed:", total)
print("Total misclassified:", mis)
print("Total sandalwood misclassified:", sandal_mis)
if sandal_idx is not None:
    preds_for_sandal = [classes[p] for (t,p) in zip(y_true,y_pred) if t==sandal_idx and p!=sandal_idx]
    print("Top 5 predicted labels for sandalwood errors:", Counter(preds_for_sandal).most_common(5))
```

```python
# Quick focused check — only Sandalwood validation images (fast)
import os, fnmatch, numpy as np
from collections import Counter
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import tensorflow as tf

PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")
MODEL_GLOB = ["*.h5","*.hdf5","*.keras"]

# find model
models = []
for root, dirs, files in os.walk(PROJECT_ROOT):
    for pattern in MODEL_GLOB:
        for fname in fnmatch.filter(files, pattern):
            models.append(os.path.join(root, fname))
models = sorted(models)
if not models:
    raise SystemExit("No model file found.")
MODEL_PATH = models[0]
print("Using model:", MODEL_PATH)

model = tf.keras.models.load_model(MODEL_PATH)
H = model.input_shape[1] or 160
W = model.input_shape[2] or 160
print("Model input:", (H,W))

sand_dir = os.path.join(VAL_DIR, "Sandalwood")
if not os.path.isdir(sand_dir):
    raise SystemExit("Sandalwood folder not found in val.")

files = [f for f in os.listdir(sand_dir) if f.lower().endswith((".jpg",".jpeg",".png"))]
print("Sandalwood files to test:", len(files))

preds = []
mis_paths = []
for i, fname in enumerate(files, 1):
    p = os.path.join(sand_dir, fname)
    img = load_img(p, target_size=(H,W))
    x = img_to_array(img)/255.0
    x = np.expand_dims(x, axis=0)
    pred_idx = int(model.predict(x, verbose=0).argmax())
    preds.append(pred_idx)
    if i % 100 == 0:
        print("Processed", i, "images...")

# map index -> class name (use val folder ordering)
classes = sorted([d for d in os.listdir(VAL_DIR) if os.path.isdir(os.path.join(VAL_DIR, d)) and not d.startswith('.')])
pred_names = [classes[p] if p < len(classes) else "UNKNOWN" for p in preds]
conf_counter = Counter(pred_names)

# results
total = len(files)
mis = sum(1 for name in pred_names if name != "Sandalwood")
print("\n=== Sandalwood-only summary ===")
print("Total sandalwood validation images:", total)
print("Total sandalwood misclassified:", mis)
print("Top 6 predicted labels for sandalwood images (most -> least):")
for cls, cnt in conf_counter.most_common(6):
    print(f"  {cls}: {cnt}")

# show examples of first 10 misclassified filenames
mis_examples = [f for f,n in zip(files, pred_names) if n != "Sandalwood"][:10]
print("\nFirst 10 misclassified sandalwood filenames (paths):")
for fn in mis_examples:
    print(" ", os.path.join(sand_dir, fn))
```
