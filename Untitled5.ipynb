```python
# Quick focused check — only Sandalwood validation images (fast)
import os, fnmatch, numpy as np
from collections import Counter
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import tensorflow as tf

PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")
MODEL_GLOB = ["*.h5","*.hdf5","*.keras"]

# find model
models = []
for root, dirs, files in os.walk(PROJECT_ROOT):
    for pattern in MODEL_GLOB:
        for fname in fnmatch.filter(files, pattern):
            models.append(os.path.join(root, fname))
models = sorted(models)
if not models:
    raise SystemExit("No model file found.")
MODEL_PATH = models[0]
print("Using model:", MODEL_PATH)

model = tf.keras.models.load_model(MODEL_PATH)
H = model.input_shape[1] or 160
W = model.input_shape[2] or 160
print("Model input:", (H,W))

sand_dir = os.path.join(VAL_DIR, "Sandalwood")
if not os.path.isdir(sand_dir):
    raise SystemExit("Sandalwood folder not found in val.")

files = [f for f in os.listdir(sand_dir) if f.lower().endswith((".jpg",".jpeg",".png"))]
print("Sandalwood files to test:", len(files))

preds = []
mis_paths = []
for i, fname in enumerate(files, 1):
    p = os.path.join(sand_dir, fname)
    img = load_img(p, target_size=(H,W))
    x = img_to_array(img)/255.0
    x = np.expand_dims(x, axis=0)
    pred_idx = int(model.predict(x, verbose=0).argmax())
    preds.append(pred_idx)
    if i % 100 == 0:
        print("Processed", i, "images...")

# map index -> class name (use val folder ordering)
classes = sorted([d for d in os.listdir(VAL_DIR) if os.path.isdir(os.path.join(VAL_DIR, d)) and not d.startswith('.')])
pred_names = [classes[p] if p < len(classes) else "UNKNOWN" for p in preds]
conf_counter = Counter(pred_names)

# results
total = len(files)
mis = sum(1 for name in pred_names if name != "Sandalwood")
print("\n=== Sandalwood-only summary ===")
print("Total sandalwood validation images:", total)
print("Total sandalwood misclassified:", mis)
print("Top 6 predicted labels for sandalwood images (most -> least):")
for cls, cnt in conf_counter.most_common(6):
    print(f"  {cls}: {cnt}")

# show examples of first 10 misclassified filenames
mis_examples = [f for f,n in zip(files, pred_names) if n != "Sandalwood"][:10]
print("\nFirst 10 misclassified sandalwood filenames (paths):")
for fn in mis_examples:
    print(" ", os.path.join(sand_dir, fn))
```

```python
# 4A — check exact duplicate files between Sandalwood and Tulsi (MD5)
import os, hashlib
PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")
sand_dir = os.path.join(VAL_DIR, "Sandalwood")
tulsi_dir = os.path.join(VAL_DIR, "Tulsi")

def md5(path):
    with open(path, "rb") as f:
        return hashlib.md5(f.read()).hexdigest()

sand_files = [f for f in os.listdir(sand_dir) if f.lower().endswith((".jpg",".jpeg",".png"))]
tulsi_files = [f for f in os.listdir(tulsi_dir) if f.lower().endswith((".jpg",".jpeg",".png"))]

sand_hashes = {md5(os.path.join(sand_dir, f)): f for f in sand_files}
tulsi_hashes = {md5(os.path.join(tulsi_dir, f)): f for f in tulsi_files}

overlap = set(sand_hashes.keys()) & set(tulsi_hashes.keys())
print("Sandalwood count:", len(sand_files))
print("Tulsi count:", len(tulsi_files))
print("Exact duplicate files between Sandalwood and Tulsi:", len(overlap))
if overlap:
    print("Examples (Sandalwood path <-> Tulsi path):")
    for h in list(overlap)[:10]:
        print(" ", os.path.join(sand_dir, sand_hashes[h]), "<->", os.path.join(tulsi_dir, tulsi_hashes[h]))
else:
    print("No exact duplicates found between the two folders.")
```

```python
# STEP 5 — Compute predicted-index -> most-common true class mapping
import os, fnmatch
from collections import Counter, defaultdict
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img, img_to_array

PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")

# find model
models = []
for root, dirs, files in os.walk(PROJECT_ROOT):
    for pat in ("*.h5", "*.hdf5", "*.keras"):
        for fn in fnmatch.filter(files, pat):
            models.append(os.path.join(root, fn))
models = sorted(models)
if not models:
    raise SystemExit("No model found.")
MODEL_PATH = models[0]
print("Using model:", MODEL_PATH)

# load model
model = tf.keras.models.load_model(MODEL_PATH)
H = model.input_shape[1] or 160
W = model.input_shape[2] or 160
print("Model input size:", (H,W))

# build class list from val folders (ignore dot-folders)
classes = sorted([d for d in os.listdir(VAL_DIR) if os.path.isdir(os.path.join(VAL_DIR, d)) and not d.startswith('.')])
print("Classes (len={}):".format(len(classes)), classes)

# iterate validation and collect mapping: predicted_index -> list of true class names
pred_to_true = defaultdict(list)
true_counts = {c: 0 for c in classes}

for i, cls in enumerate(classes):
    cls_folder = os.path.join(VAL_DIR, cls)
    imgs = [f for f in os.listdir(cls_folder) if f.lower().endswith((".jpg",".jpeg",".png"))]
    for fname in imgs:
        p = os.path.join(cls_folder, fname)
        img = load_img(p, target_size=(H,W))
        x = img_to_array(img)/255.0
        x = np.expand_dims(x, axis=0)
        preds = model.predict(x, verbose=0)
        pred_idx = int(np.argmax(preds, axis=-1)[0])
        pred_to_true[pred_idx].append(cls)
        true_counts[cls] += 1

# Summarize mapping: for each predicted index, show top 3 true classes that mapped to it
print("\n=== Predicted-index -> most common TRUE classes (top 3) ===")
mapping = {}
for pred_idx in sorted(pred_to_true.keys()):
    ctr = Counter(pred_to_true[pred_idx])
    top = ctr.most_common(3)
    total = sum(ctr.values())
    mapping[pred_idx] = ctr
    print(f"Predicted index {pred_idx:2d}  (total assigned: {total:4d})  top true classes -> {top}")

# Show which predicted index is the model using most for Sandalwood
if "Sandalwood" in classes:
    sand_idx = classes.index("Sandalwood")
    sand_pred_counts = Counter()
    sand_folder = os.path.join(VAL_DIR, "Sandalwood")
    sand_imgs = [f for f in os.listdir(sand_folder) if f.lower().endswith((".jpg",".jpeg",".png"))]
    for fname in sand_imgs:
        p = os.path.join(sand_folder, fname)
        img = load_img(p, target_size=(H,W))
        x = img_to_array(img)/255.0
        x = np.expand_dims(x, axis=0)
        pred_idx = int(model.predict(x, verbose=0).argmax())
        sand_pred_counts[pred_idx] += 1
    print("\n=== Sandalwood predicted-index distribution ===")
    for idx, cnt in sand_pred_counts.most_common():
        print(f"  Predicted index {idx:2d} => {cnt} images")
else:
    print("Sandalwood not found among classes (spelling/case mismatch).")

# Quick helper: show if any predicted index maps primarily to 'Tulsi'
tulsi_pred_idx = None
for idx, ctr in mapping.items():
    if ctr:
        top_class, top_count = ctr.most_common(1)[0]
        if top_class == "Tulsi":
            tulsi_pred_idx = idx
            print(f"\nNote: Predicted index {idx} is dominated by true class 'Tulsi' (count {top_count})")

print("\nFinished Step 5 mapping check.")
```

```python
# Fix: build predicted-index -> majority true-class mapping, remap preds, save report
import os, fnmatch, json, csv
import numpy as np
import tensorflow as tf
from collections import defaultdict, Counter
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt

PROJECT_ROOT = r"C:/MedicinalPlantApp"
VAL_DIR = os.path.join(PROJECT_ROOT, "data", "val")
OUT_MAP = os.path.join(PROJECT_ROOT, "pred_index_to_class.json")
OUT_REPORT = os.path.join(PROJECT_ROOT, "classification_report_remapped.csv")
OUT_CM = os.path.join(PROJECT_ROOT, "confusion_matrix_remapped.csv")
OUT_CM_PNG = os.path.join(PROJECT_ROOT, "confusion_matrix_remapped.png")

# find model
models = []
for root, dirs, files in os.walk(PROJECT_ROOT):
    for pat in ("*.h5","*.hdf5","*.keras"):
        for f in fnmatch.filter(files, pat):
            models.append(os.path.join(root, f))
models = sorted(models)
if not models:
    raise SystemExit("No model file found.")
MODEL_PATH = models[0]
print("Using model:", MODEL_PATH)
model = tf.keras.models.load_model(MODEL_PATH)
H = model.input_shape[1] or 160
W = model.input_shape[2] or 160
print("Model input:", (H,W))

# build class list
classes = sorted([d for d in os.listdir(VAL_DIR) if os.path.isdir(os.path.join(VAL_DIR,d)) and not d.startswith('.')])
print("Classes:", classes)

# collect predicted_index -> true-class list
pred_to_true = defaultdict(list)
y_true_idx = []
y_pred_idx = []
paths = []

for i, cls in enumerate(classes):
    folder = os.path.join(VAL_DIR, cls)
    imgs = [f for f in os.listdir(folder) if f.lower().endswith((".jpg",".jpeg",".png"))]
    for fname in imgs:
        p = os.path.join(folder, fname)
        try:
            img = tf.keras.preprocessing.image.load_img(p, target_size=(H,W))
            x = tf.keras.preprocessing.image.img_to_array(img)/255.0
            x = np.expand_dims(x, axis=0)
            preds = model.predict(x, verbose=0)
            pred_idx = int(np.argmax(preds, axis=-1)[0])
        except Exception as e:
            print("Error processing:", p, "->", e)
            continue
        pred_to_true[pred_idx].append(cls)
        y_true_idx.append(i)
        y_pred_idx.append(pred_idx)
        paths.append(p)

# make mapping: predicted index -> majority true class
pred_map = {}
for pred_idx, trues in pred_to_true.items():
    ctr = Counter(trues)
    most_common_class, most_count = ctr.most_common(1)[0]
    pred_map[int(pred_idx)] = most_common_class

# for predicted indices not present (rare), map to UNKNOWN
num_classes = max(len(classes), max(pred_map.keys())+1)
for k in range(num_classes):
    if k not in pred_map:
        pred_map[k] = "UNKNOWN"

# save mapping
with open(OUT_MAP, "w", encoding="utf-8") as f:
    json.dump(pred_map, f, indent=2)
print("\nSaved predicted-index->class map to:", OUT_MAP)
print("Example mapping (first 10):", dict(list(pred_map.items())[:10]))

# Remap predictions to class names using mapping
y_true = np.array(y_true_idx)
y_pred_mapped = np.array([pred_map.get(int(p), "UNKNOWN") for p in y_pred_idx])
y_true_names = np.array([classes[i] for i in y_true_idx])

# classification report (remapped)
report = classification_report(y_true_names, y_pred_mapped, labels=classes, output_dict=True, zero_division=0)
print("\n=== Remapped classification report summary (showing Sandalwood entry) ===")
if "Sandalwood" in report:
    s = report["Sandalwood"]
    print(f"Sandalwood  precision: {s['precision']:.4f}  recall: {s['recall']:.4f}  f1: {s['f1-score']:.4f}  support: {int(s['support'])}")
else:
    print("Sandalwood not in report (unexpected).")

# save report CSV
with open(OUT_REPORT, "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["class","precision","recall","f1-score","support"])
    for cls in classes:
        pr = report.get(cls, {"precision":0,"recall":0,"f1-score":0,"support":0})
        writer.writerow([cls, pr["precision"], pr["recall"], pr["f1-score"], int(pr["support"])])
print("Saved remapped classification report to:", OUT_REPORT)

# build remapped confusion matrix (rows=true classes, cols=predicted mapped classes in same order)
mapped_preds_idx = [classes.index(pred) if pred in classes else -1 for pred in y_pred_mapped]
# if any -1 (UNKNOWN), temporarily append UNKNOWN column
if -1 in mapped_preds_idx:
    print("Some mapped predictions were UNKNOWN; they will be included as an extra column in confusion csv.")

cm = confusion_matrix(y_true_names, y_pred_mapped, labels=classes)
# save cm as csv
with open(OUT_CM, "w", newline='', encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["true\\pred"] + classes)
    for i, row in enumerate(cm):
        writer.writerow([classes[i]] + row.tolist())
print("Saved remapped confusion matrix to:", OUT_CM)

# plot normalized heatmap
cm_norm = cm.astype('float') / (cm.sum(axis=1, keepdims=True) + 1e-8)
fig, ax = plt.subplots(figsize=(12,10))
im = ax.imshow(cm_norm, interpolation='nearest', aspect='auto')
ax.set_xticks(range(len(classes))); ax.set_xticklabels(classes, rotation=90, fontsize=8)
ax.set_yticks(range(len(classes))); ax.set_yticklabels(classes, fontsize=8)
plt.colorbar(im, ax=ax, fraction=0.03)
plt.title("Confusion matrix (remapped preds) — normalized by true class")
plt.tight_layout()
plt.savefig(OUT_CM_PNG, dpi=150)
print("Saved confusion matrix image to:", OUT_CM_PNG)
plt.show()

print("\nDone. Look at", OUT_REPORT, "and", OUT_CM, "and check Sandalwood row in the remapped report.")
```
